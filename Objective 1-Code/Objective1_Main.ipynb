# **Correlation Analysis**

```python
GHG_cor = GHG.drop(columns=['Country', 'Year']) #dropping non-numeric and unwanted columns
GHG_cor

GHG_cor = GHG_cor.rename(columns={'Greenhouse gas emissions from agriculture': 'Agriculture', 'Greenhouse gas emissions from land use change and forestry':'land use change and forestry','Greenhouse gas emissions from waste': 'Waste', 'Greenhouse gas emissions from buildings':'Buildings','Greenhouse gas emissions from industry':'Industry','Greenhouse gas emissions from manufacturing and construction':'Manufacturing and Construction','Greenhouse gas emissions from transport':'Transport','Greenhouse gas emissions from electricity and heat':'Electricity and Heat','Fugitive emissions of greenhouse gases from energy production':'Energy','Greenhouse gas emissions from other fuel combustion':'Fuel Combustion','Greenhouse gas emissions from bunker fuels':'Bunker Fuels'})
```

```python
Correlation_GHG = GHG_cor.corr().round(2) #Correlation matrix for GHG numeric
Correlation_GHG

print(Correlation_GHG.to_string())
```

```python
# Create heatmap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 10))  # Adjust figure size as needed

# Use tril_indices instead of tril_indices_n if using older NumPy version
mask = np.zeros_like(Correlation_GHG)
mask[np.triu_indices(Correlation_GHG.shape[0])] = True

# Use the correct function name for diverging colormaps in Seaborn
cmap = sns.color_palette("RdBu_r", as_cmap=True)  # Define diverging colormap
cmap.set_bad('white')

sns.heatmap(Correlation_GHG, annot=True, cmap=cmap, fmt=".1f", linewidths=.5, mask=mask)
plt.title("Correlation Heatmap for GHG Emissions")
plt.show()
```
