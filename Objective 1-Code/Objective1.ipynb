```python
GHG_numeric = GHG.drop(columns=['Code', 'Entity', 'Year'])
```

```python
Corr_GHG = GHG_numeric.corr()
Corr_GHG #Correlation matrix for GHG numeric
```

```python
# Create heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(Corr_GHG, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap for GHG Emissions")
plt.show()
```
